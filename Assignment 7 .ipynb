```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('matches - matches.csv')
```

```python
## data.player_of_match  -> access the player_of_match of the match column 
### vlaue_counts() -> counts the player and prize it won
### head() ->this method returns the top 3 data 
```

```python
data
```

```python
mask1=data.city=='Mumbai'
data[mask1].player_of_match.value_counts().head(3)
```

```python
def top_batsman(city1):
    mask2=data.city==city1
    return data[mask2].player_of_match.value_counts().head(3).plot(kind="pie",figsize=(8,8))
    
    
```

```python
top_batsman("Mumbai")
```

```python
top_batsman('Hyderabad')
```
